```python
def category(num):
    import ast
    
    if ast.literal_eval(num[:-3]) > 0 and ast.literal_eval(num[:-3]) <=3:
        return "0-3"
    elif ast.literal_eval(num[:-3]) > 3 and ast.literal_eval(num[:-3]) <=6:
        return "3-6"
    elif ast.literal_eval(num[:-3]) > 6 and ast.literal_eval(num[:-3]) <=9:
        return "6-9"
    elif ast.literal_eval(num[:-3]) > 9 and ast.literal_eval(num[:-3]) <=12:
        return "9-12"
    elif ast.literal_eval(num[:-3]) > 12 and ast.literal_eval(num[:-3]) <=15:
        return "12-15"
    elif ast.literal_eval(num[:-3]) > 15 and ast.literal_eval(num[:-3]) <=18:
        return "15-18"
    elif ast.literal_eval(num[:-3]) > 18 and ast.literal_eval(num[:-3]) <=21:
        return "18-21"
    elif ast.literal_eval(num[:-3]) > 21:
        return "21-24"
```

```python
def process_data(filename):
    
    import time
    import ast
    import pandas as pd
    import numpy as np
    
    df = pd.read_csv(filename)
    print (df.columns)
    try:
        print(df.isna().sum())
        df.dropna(inplace=True)
        df.drop(['Modified_Unit_ID','Sync_Id','Sync_Id_Old'], axis=1, inplace=True)
        df.drop(['Action_Flag'], axis=1, inplace=True)
        df.drop_duplicates(subset=['Registration_Number'])
        time_frame = df['DEDT'].apply(lambda x : str(x[-5:]))
        df['Time'] = time_frame
        df['Date'] = df['DEDT'].apply(lambda x:str(x[:-5]))
        df.drop_duplicates(subset=['UserID'], inplace = True)
    except Exception as e:
        print(e)
        pass
    
    print('Save returned df as a file')
    return df
```
